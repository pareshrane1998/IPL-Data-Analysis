```python
# Players sold and uinsold using bar graph
import seaborn as sns
plt.figure(figsize=(10,5))
fig = sns.countplot(ipl['status'],palette = ['orange','pink'])
plt.xlabel('Sold or Unsold Players')
plt.ylabel('Number of Players')
plt.title('Sold vs Unsold Players',fontsize=15)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(),'.0f'),(p.get_x() + p.get_width()/2.,p.get_height()),ha='center',va='center',
                xytext=(0,4),textcoords='offset points')
```

```python
ipl['base_price'].replace('Retained',0,inplace=True)
```
